# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

# Calculate Correlation

```python
import pandas as pd

# Load your data
base_dir="/home/jovyan/work/pv-digital-twin"
df = pd.read_csv(f"{base_dir}/data/Jingxing-Station05/final/data.csv")

# Calculate the Pearson correlation coefficient
correlation = df['irradiance'].corr(df['active_power'], method='pearson')

print(f"Correlation between Irradiance and Active Power: {correlation:.4f}")
```

# Plot Function

```python
def plot_monthly(df, metric):
  # Make sure datetime is parsed
  df['collect_time'] = pd.to_datetime(df['collect_time'])
  df['month'] = df['collect_time'].dt.month  # extract month

  # Set up the figure with 12 subplots
  fig, axes = plt.subplots(4, 3, figsize=(18, 12), sharey=True)
  fig.suptitle(f'{metric}, by Month', fontsize=16)

  # Plot each month in a separate subplot
  for month in range(1, 13):
      ax = axes[(month-1)//3, (month-1)%3]  # get subplot position
      monthly_df = df[df['month'] == month].copy()

      monthly_df[metric] = pd.to_numeric(monthly_df[metric], errors='coerce')
      monthly_df = monthly_df.dropna(subset=['collect_time', metric])

      if not monthly_df.empty:
          ax.plot(monthly_df['collect_time'], monthly_df[metric], linewidth=0.5)

      # Show only day of the month on x-axis
      ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
      ax.xaxis.set_major_locator(mdates.DayLocator())

      ax.set_title(f'Month {month}')
      ax.set_xlabel('Day')
      ax.set_ylabel(metric)
      ax.grid(True)

  for ax in axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)    # or try 60
        label.set_fontsize(8)     # smaller font
  plt.tight_layout(rect=[0, 0, 1, 0.96])  # leave space for main title
  plt.show()
```

# Plot Dataset

```python
# Sampling frequency setup
samples_per_day = 288
num_days = 10

# Filter for 10 days
df_10_days = df.iloc[:samples_per_day * num_days]

# Plot
fig, axes = plt.subplots(num_days, 1, figsize=(14, 25), sharey=True)

for i in range(num_days):
    start = i * samples_per_day
    end = (i + 1) * samples_per_day
    day_data = df_10_days.iloc[start:end]
    
    axes[i].plot(day_data['collect_time'], day_data['irradiance'], label='Irradiance', linewidth=1.5)
    axes[i].plot(day_data['collect_time'], day_data['active_power'], label='Active Power', linewidth=1.5)
    axes[i].set_title(f'Day {i+1}')
    axes[i].legend()

plt.tight_layout()
plt.xlabel('Time')
plt.show()
```
